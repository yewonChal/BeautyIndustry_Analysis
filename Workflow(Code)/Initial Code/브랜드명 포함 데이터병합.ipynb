```python
import pandas as pd

# 파일 불러오기
df_DrG = pd.read_csv('review_data(닥터지).csv', encoding='cp949')
df_manyo = pd.read_csv('review_data(마녀공장).csv', encoding='cp949')
df_Abib  = pd.read_csv('review_data(아비브).csv', encoding='cp949')
df_aestura  = pd.read_csv('review_data(에스트라).csv', encoding='cp949')
df_AHC  = pd.read_csv('review_data(AHC).csv', encoding='cp949')
df_CNP  = pd.read_csv('review_data(CNP).csv', encoding='cp949')

df_manyo = df_manyo.rename(columns={'Brand': 'brand'})

df_all = pd.concat([df_DrG, df_manyo, df_Abib, df_aestura, df_AHC, df_CNP], ignore_index=True)

df_all.to_csv('data_all.csv', index=False, encoding='cp949')
```
