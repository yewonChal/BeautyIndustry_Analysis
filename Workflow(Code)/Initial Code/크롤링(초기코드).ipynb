```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.oliveyoung.co.kr/store/goods/getGoodsDetail.do?goodsNo=A000000207229&dispCatNo=90000010009&trackingCd=Best_Sellingbest&t_page=랭킹&t_click=판매랭킹_스킨케어_상품상세&t_number=1"

response = requests.get(url)
response.raise_for_status()

soup = BeautifulSoup(response.text, 'html.parser')

df_review_cos1 = pd.DataFrame(columns=[
    'date', 'rate', 'id', 'skin_type', 'select_1_title', 'select_1_content',
    'select_2_title', 'select_2_content', 'select_3_title', 'select_3_content', 'txt'
])

reviews = soup.select('#gdasList > li')

for review in reviews:
    try:
        date = review.select_one('div.info > div > p > span.date').text.strip()
        rate = review.select_one('div.review_cont > div.score_area > span.review_point > span').text.strip()
        user_id = review.select_one('div.info > div > p.info_user > a.id').text.strip()
        skin_type = review.select_one('div.review_cont > div.poll_sample > dl:nth-child(1) > dd > span').text.strip()
        select_1_title = review.select_one('div.review_cont > div.poll_sample > dl:nth-child(1) > dt').text.strip()
        select_1_content = review.select_one('div.review_cont > div.poll_sample > dl:nth-child(1) > dd > span').text.strip()
        select_2_title = review.select_one('div.review_cont > div.poll_sample > dl:nth-child(2) > dt').text.strip()
        select_2_content = review.select_one('div.review_cont > div.poll_sample > dl:nth-child(2) > dd > span').text.strip()
        select_3_title = review.select_one('div.review_cont > div.poll_sample > dl:nth-child(3) > dt').text.strip()
        select_3_content = review.select_one('div.review_cont > div.poll_sample > dl:nth-child(3) > dd > span').text.strip()
        txt = review.select_one('div.txt_inner').get_text(separator="\n").strip()

        df_review_cos1 = df_review_cos1.append({
            'date': date,
            'rate': rate,
            'id': user_id,
            'skin_type': skin_type,
            'select_1_title': select_1_title,
            'select_1_content': select_1_content,
            'select_2_title': select_2_title,
            'select_2_content': select_2_content,
            'select_3_title': select_3_title,
            'select_3_content': select_3_content,
            'txt': txt
        }, ignore_index=True)
    except AttributeError as e:
        print(f"Failed to scrape review: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

df_review_cos1.to_csv('/Users/jeon-yewon/Desktop/test/review_data.csv', index=False)

print("크롤링 완료 및 데이터 저장")
```
