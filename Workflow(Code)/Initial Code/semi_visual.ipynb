```python
def remove_newline(text):
    return re.sub(r'\n', '', text)

review['리뷰'] = review['리뷰'].apply(remove_newline)
```

```python
import re

# 평점 중 점수 추출
def score_conversion(text):
    score_match = re.search(r'(\d+)점$', text) # 숫자 다음 '점'이 있는 패턴 찾기, ex) 5점만점에 4점
    if score_match:
        return score_match.group(1) # 추출된 숫자 반환(score_match에서 괄호로 묶인 첫 번째 그룹(숫자) 반환)
    else:
        return None

review['평점'] = review['평점'].apply(score_conversion)
```

```python
review_sentences = []
review_sentences.extend(list(review['리뷰'].values))
review_sentences[:5]
```

```python
from string import punctuation

def repreprocessing(review_sentence):  # 구두점 제거
    return ''.join(word for word in review_sentence if word not in punctuation)

# def repreprocessing(review_sentence):
#     return ''.join(word for word in review_sentence if word not in punctuation or word == '+')
```

```python
# 이모티콘 제거
import re

def remove_emojis(text):
    emoji_pattern = re.compile("\U0001F600-\U0001F64F", flags=re.UNICODE)

    return emoji_pattern.sub(r'', text)

review['리뷰'] = review['리뷰'].apply(remove_emojis)
```

```python
# 자음 반복 제거.
def cleanText2(reviews):
    text = re.sub('[ㄱㄴㄷㄹㅁㅂㅅㅇㅈㅊㅋㅌㅍㅎㅃㅉㄸㄲㅆㅛㅕㅑㅐㅔㅗㅓㅏㅣㅜㅠㅡ]', '', reviews)
    return text

review['리뷰'] = review['리뷰'].apply(cleanText2)
```

```python
# 무적의 코드. 특수문자 모두 제거
def cleanText(text):
    text = re.sub('[-=,#/\?:;^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》❤️♥️♡]', '', text)
    return text

review['리뷰'] = review['리뷰'].apply(cleanText)
```
